```python
import numpy as np
import pandas as pd
```

### import dataset

```python
data = pd.read_csv("./home-data-for-ml-course/train.csv", index_col="Id")
data.info()
```

### select feature

```python
features=['LotArea', "YearBuilt", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
X = data[features]
y = data["SalePrice"]
```

### Split data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)
```

### Training Machine Learning model

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_preds=dt_model.predict(X_test.head(10))
y_preds
```

```python
# DataFrame to compare y_test and y_preds
compare = pd.DataFrame({"y_test":y_test.head(10), "y_preds": y_preds})
compare
```

### Training by RandomForestRegression

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
rf_model = RandomForestRegressor(random_state=1)
```

```python
rf_model.fit(X_train, y_train)
```

```python
preds_rf_value = rf_model.predict(X_test)
# preds_rf_value
```
