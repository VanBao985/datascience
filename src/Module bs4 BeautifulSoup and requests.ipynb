# ***Module bs4: BeautifulSoup and Module requests***

BeautifulSoup là một thư viện Python dùng để lấy dữ liệu ra khỏi các file HTML và XML. Nó hoạt động cùng với các parser (trình phân tích cú pháp) cung cấp cho bạn các cách để điều hướng, tìm kiếm và chỉnh sửa trong parse tree (cây phân tích được tạo từ parser). Nhờ các parser này nó đã giúp các lập trình viên tiết kiệm được nhiều giờ làm việc.

```python
from bs4 import BeautifulSoup
import requests
```

1. Làm quen với BeautifulSoup

```python
page = requests.get("https://vnexpress.net/the-gioi")
print(page)
page.text
```

```python
soup = BeautifulSoup(page.text, "html.parser")
soup.prettify()
```

```python
print(soup.title)
print(soup.p)     #find the first!
print(soup.a)
```

```python
soup.find_all('p')
```

```python
for link in soup.find_all('a'):
  print(link.get('href')) if link.get('href') != None else print(end="")
```

```python
soup.find_all('p')[5].get_text()
```

```python
for i in range(10):
  print(soup.find_all('p')[i].get_text())
```

2. Lam quen voi requests

```python
import requests
page = requests.get("https://api.github.com")
page
```

```python
if page.status_code == 200:
    print("Success!")
elif page.status_code == 404:
    print("Not Found.")
page.status_code
```

```python
page.encoding="utf-8"
page.text
```

```python
page_dict = page.json()
print(page_dict['followers_url'])
# page_dict
```

```python
# Search GitHub's repositories for popular Python projects
response = requests.get(
    "https://api.github.com/search/repositories",
    params={"q": "language:python", "sort": "stars", "order": "desc"},
)
response_json = response.json()
for repo in response_json['items'][:4]:
  print(f"Name: {repo['name']}")
  print(f"Description: {repo['description']}")
  print(f"Stars: {repo['stargazers_count']}")
  print()
```
